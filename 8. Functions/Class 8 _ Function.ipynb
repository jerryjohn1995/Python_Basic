**Q:** **Consider a list**

`l = [2,3,4,5,6,7,8]`

Without using the inbuilt function find the sum of the list.

**1. Without using function.**

**2.** **With function. (Return the result)**

````python
#1. Without using function

l = [2,3,4,5,6,7,8]
a=0

for i in l:
  a = a + i

print(f"The final sum of the list is {a}")
````

````python
def sum_list():
  l = [2,3,4,5,6,7,8]
  a=0

  for i in l:
    a = a + i

  return a

result = sum_list()

print(f"The result of the sum_list function is {result}")
````

**3. If the list is added with a string element, and we only need to add the numbers in it,**

`l = [2,3,4,5,6,7,8,9,"jerry"]`



````python
def sum_list():
  l = [2,3,4,5,6,7,8,9,"jerry"]
  a=0

  for i in l:
    if(type(i) == int):
      a = a + i

  return a

result = sum_list()

print(f"The result of the sum_list function is {result}")
````

# **1.4: Passing value at function call**

````python
#Function definition
def sum_list(l):
  a=0
  for i in l:
    if(type(i) == int):
      a = a + i

  return a

#Main body
l = [2,3,4,5,6,7,8,9,"jerry",10]
result = sum_list(l)

print(f"The result of the sum_list function is {result}")
````

**Q: Given a list, try to create a new list with only the string elements.**

`l = [1, 2, "hai", 3, 4, 5, 6, 7, "jerry", 8, "john", 9]`

````python
l = [1, 2, "hai", 3, 4, 5, 6, 7, "jerry", 8, "john", 9]
````

````python
#Method 1 (Using function, the easy way)

def only_strings(l):
  string_list = []  #Empty list for creating the new list
  for i in l:
    if type(i) == str:
      string_list.append(i)
  return string_list

result = only_strings(l)
print(f"The list with only strings {result}")
````

````python
#Method 2 (The difficult way)

new_list_only_strings = [i for i in l if type(i) == str]
print(new_list_only_strings)
````
