# **1.   Write a function which will try to find out len of a string without using an inbuilt len function?**


```python
def string_len(sentence):
  """
  This is a program to count the length of a function without using the inbuilt len feunction
  """
  count = 0
  for i in sentence:
    count = count + 1
  return count

sentence = "jerryjohn"

length = string_len(sentence)
print(f"The lenghth of the string is {length}")
```

# **2.   Write a function which will be able to print an index of list element without using an index function.**

```python
l = [1, 2, 3, 4, 5, "sudh", "jerry"]


def index_list(l):
  """
  Function to print an index of list element without using index function
  """
  for i in range(0, len(l)):
    print(l[i],"-->", i)

index_list(l)
```

# **3.   Write a function which will be able to print an ip address of your system.**


```python
import socket

def get_ip():
  """
  Function to print an ip address
  """
  ip = socket.gethostbyname(socket.gethostname())
  print(ip)

get_ip()
```

# **5.   Write a function which will take input as a list with any kind of numerical value and give an output as a multiplication of all the numeric data.**
`l = [3.5, 6.56, 4, 4, "sugh", "jerry", "john"]`

```python
def mul_numeric(inp_list):
  """
  Function to multiply the numeric elements of a list
  """
  result = 1
  for i in inp_list:
    if( type(i) == int or type(i) == float):
      result = result * i

  return result

l = [3.5, 6.56, 4, 4, "sugh", "jerry", "john"]

result = mul_numeric(l)
print(f"The result is {result}")
```

# **6.   Write a function which will be able to read all the mails.**

```python
import smtplib
import time
import imaplib
import email
import traceback 
ORG_EMAIL = "@gmail.com" 
FROM_EMAIL = 'sskumar9876@gmail.com'
FROM_PWD = 'rlplfdcsoiqruagn'
SMTP_SERVER = "imap.gmail.com" 
SMTP_PORT = 993
imaplib._MAXLINE = 400000000

def read_email_from_gmail():
    try:
        mail = imaplib.IMAP4_SSL(SMTP_SERVER)
        mail.login(FROM_EMAIL,FROM_PWD)
        mail.select('inbox')

        data = mail.search(None, 'ALL')
        mail_ids = data[1]
        id_list = mail_ids[0].split()   
        first_email_id = int(id_list[0])
        latest_email_id = int(id_list[-1])

        for i in range(latest_email_id,first_email_id, -1):
            data = mail.fetch(str(i), '(RFC822)' )
            for response_part in data:
                arr = response_part[0]
                if isinstance(arr, tuple):
                    msg = email.message_from_string(str(arr[1],'utf-8'))
                    email_subject = msg['subject']
                    email_from = msg['from']
                    print('From : ' + email_from + '\n')
                    print('Subject : ' + email_subject + '\n')

    except Exception as e:
        traceback.print_exc() 
        print(str(e))

read_email_from_gmail()
```
